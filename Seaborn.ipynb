#1 - Plots de distribuição

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=300)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x='total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
sns.jointplot(x='total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.pairplot(tips, hue = "sex") # palette="coolwarm"
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

# 2 -Plots Categóricos

# 3 - plots de regressão

```python
tips
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s': 100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col= 'time', row='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s': 100}, aspect=1.6, size=8)
```

# 4 - Plots Matriciais

```python
flights = sns.load_dataset('flights')
flights
```

```python
df_flights = flights.pivot_table(index='year', columns='month', values='passengers')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(20, 8))
sns.heatmap(df_flights, ax=ax, annot = True, fmt = '.0f', linewidths = 1, cmap='magma')
```

```python
sns.clustermap(df_flights)
```
