## Orientações à objetos no matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #esquerda, inferior, largura, altura

axes.plot(x, y, 'blue');
axes.set_xlabel('Label X')
axes.set_ylabel('Label y')
axes.set_title('Título')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #esquerda, inferior, largura, altura
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes.plot(x, y, 'blue');
axes.set_xlabel('Label X')
axes.set_ylabel('Label y')
axes.set_title('Título')

axes2.plot(y, x, 'red');
axes2.set_xlabel('Label X2')
axes2.set_ylabel('Label y2')
axes2.set_title('Título 2')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols= 2) 

for axis in ax:
  axis.plot(x, y, 'r')
  axis.set_title("Titulo")
```

```python
fig, ax = plt.subplots(nrows = 1, ncols= 2) 

ax[0].plot(x, y, 'red')
ax[1].plot(x, y, 'green')
```

```python
fig, ax = plt.subplots(nrows = 1, ncols= 2, figsize=(30,8)) 

for axis in ax:
  axis.plot(x, y, 'r')
```

## Customização

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, 'b--' ,label = "x**2", alpha = 0.5)
ax.plot(x, x**3, 'r--',label = "x**3")

ax.set_title("Título")
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo y')

ax.legend(loc= 6)
```

```python
fig, axes = plt.subplots(1, 3, figsize = (12, 4))

axes[0].plot(x, x**2)
axes[0].plot(x, x**3)
axes[0].set_title('Intervalo de eixos padrão')

axes[1].plot(x, x**2)
axes[1].plot(x, x**3)
axes[1].axis('tight')
axes[1].set_title('Eixos apertados')

axes[2].plot(x, x**2)
axes[2].plot(x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2,5])
axes[2].set_title('Intervalo de eixos com zoom')
```
